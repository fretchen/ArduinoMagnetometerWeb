```python
import socket
import sys
import time
import serial
import datetime as dt
import numpy as np
import h5py

HOST = ''   # Symbolic name meaning all available interfaces
PORT = 8899 # Arbitrary non-privileged port
 

try:
    ser = serial.Serial('COM32', 9600, timeout = 1)
except serial.SerialException:
    s.close()
    ser.close()
    ser = serial.Serial('COM32', 9600, timeout = 1)

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
print 'Socket created'

try:
    s.bind((HOST, PORT))
except socket.error , msg:
    print 'Bind failed. Error Code : ' + str(msg[0]) + ' Message ' + msg[1]
    sys.exit()
     
print 'Socket bind complete'

s.listen(5)
print 'Socket now listening'


    
    
print 'Serial COM32 is now listening'
while True:
    #wait to accept a connection - blocking call
    conn, addr = s.accept()
 
    #display client information
    print 'Connected with ' + addr[0] + ':' + str(addr[1])

    #now get stuff from the client
    data = conn.recv(1024)
    reply = 'OK...' + data
    
    ser.flushInput();
    line = ser.readline();
    ard_str = line.decode(encoding='windows-1252');
    try:
        vals = np.array(ard_str.split(','), dtype =float)    
        fname = 'N:\\' + data[3:]
        time.sleep(5)
        f = h5py.File(fname, "a")
        params = f['globals']
        params.attrs['T_Verr'] = vals[0]
        params.attrs['T_Vmeas'] = vals[1]
        params.attrs['T_Vinp'] = vals[2]
        f.close()
        print vals
        print reply
    except (ValueError, IndexError) as e:
        print 'Convert failed for string: ' + ard_str
        

    # if data is zero this means that the connection was lost and we can actually close our server
    if not data: 
        break
    conn.sendall(reply)
    time.sleep(0.2)
    conn.close()
```
